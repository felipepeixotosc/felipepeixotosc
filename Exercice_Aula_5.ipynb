## 1. Importando arquivo e entendendo os dados

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("topyoutube 3.csv")
df.head(1000)
```

```python
df.info()
```

## 2. Tratar as colunas sem informação adicionando o texto “Não informado”.

```python
df['Total Views'] = df['Total Views'].str.replace(',', '').astype(float)
```

```python
df['100M'].fillna('não informado', inplace=True)
```

```python
df.info()
```

## 3.Formatar a coluna de total de inscritos multiplicando por 100 milhões

```python
df['Total Views'] = df['Total Views']/100000000
```

## 4. Formatar coluna “AVG” para 2 casas decimais depois da vírgula.

```python
df['Avg'] = df['Avg'].map(lambda x: '{:.2f}'.format(x))
```

## 5. Mostrar os top 10 usuários.

```python
top_10_artist = df.sort_values(by='Total Views', ascending=False)
```

```python
top_10_artist.head(10)
```

## 6. Mostrar primeiros 100 usuários por nome decrescente

```python
pd.set_option('display.max_rows', 100) 
top_100_artist = df.sort_values(by='Artist' , ascending=False)
```
